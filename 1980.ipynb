```python
# Produit 1 : iPhone
id1 = 1
produit1 = "Iphone"
PrixUV1 = 2500      # Prix Unitaire de Vente
PrixUA1 = 2300      # Prix Unitaire d'Achat
promis1 = 10
quantite1 = 10
```

```python
# Produit 2 : Samsung
id2 = 2
produit2 = "Samsung"
PrixUV2 = 1800
PrixUA2 = 1500
promis2 = 15
quantite2 = 15
```

```python
# Calcul du chiffre d'affaire (CA)
CA1 = PrixUV1 * quantite1
CA2 = PrixUV2 * quantite2
CA_total = CA1 + CA2
```

```python
# Calcul du coût de revient (CR)
CR1 = PrixUA1 * quantite1
CR2 = PrixUA2 * quantite2
CR_total = CR1 + CR2
```

```python
# Calcul du bénéfice (CA - CR)
benefice1 = CA1 - CR1
benefice2 = CA2 - CR2
benefice_total = benefice1 + benefice2
```

```python
# Affichage des résultats
print("---- Résultats par produit ----")
print("Produit:", produit1)
print("CA =", CA1, "CR =", CR1, "Bénéfice =", benefice1)
print("Produit:", produit2)
print("CA =", CA2, "CR =", CR2, "Bénéfice =", benefice2)

print("\n---- Totaux ----")
print("CA total =", CA_total)
print("CR total =", CR_total)
print("Bénéfice total =", benefice_total)
```
