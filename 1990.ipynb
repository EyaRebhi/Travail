```python
import sqlite3
```

```python
# =======================================
# Connexion à la base de données (en mémoire)
# =======================================
conn = sqlite3.connect(":memory:")
curseur = conn.cursor()

# =======================================
```

```python
# 1. Création de la table Produit
# =======================================
curseur.execute("""
CREATE TABLE Produit (
    ID INTEGER PRIMARY KEY,
    Nom TEXT,
    PrixUV REAL,
    PrixUA REAL,
    Promise INTEGER,
    Quantite INTEGER
)
""")
```

```python
# =======================================
# 2. Insertion des données
# =======================================
curseur.execute("INSERT INTO Produit VALUES (1, 'Iphone', 2500, 2300, 10, 10)")
curseur.execute("INSERT INTO Produit VALUES (2, 'Samsung', 1800, 1500, 15, 15)")

# =======================================
```

```python
# 3. Calcul du CA, CR et bénéfice via SQL
# =======================================
curseur.execute("""
SELECT
    Nom,
    PrixUV * Quantite AS CA,
    PrixUA * Quantite AS CR,
    (PrixUV * Quantite) - (PrixUA * Quantite) AS Benefice
FROM Produit
""")

resultats = curseur.fetchall()
```

```python
# Affichage des résultats
for ligne in resultats:
    print("Produit:", ligne[0])
    print("  Chiffre d'Affaires (CA):", ligne[1])
    print("  Coût de Revient (CR):", ligne[2])
    print("  Bénéfice:", ligne[3])
    print("-----------------------------")

# =======================================
```

```python
# =======================================
# 4. Totaux globaux
# =======================================
curseur.execute("""
SELECT
    SUM(PrixUV * Quantite) AS Total_CA,
    SUM(PrixUA * Quantite) AS Total_CR,
    SUM((PrixUV * Quantite) - (PrixUA * Quantite)) AS Total_Benefice
FROM Produit
""")

total = curseur.fetchone()
print("TOTAL Chiffre d'Affaires:", total[0])
print("TOTAL Coût de Revient:", total[1])
print("TOTAL Bénéfice:", total[2])

conn.close()
```
